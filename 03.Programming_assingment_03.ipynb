### Q1. Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python
num=int(input('Enter a number : '))
if num>0:
    print('Enter {} is Positive '.format(num))
elif num<0:
    print('Enter {} is Negative '.format(num))
else:
    print('Enter number is Zero')
```

### Q2. Write a Python Program to Check if a Number is Odd or Even?

```python
num=int(input('Enter a number : '))

if num%2==0:
    print('Enter number is Even')
else:
    print('Enter number is Odd')
    
```

### Q3. Write a Python Program to Check Leap Year?

```python
def check_year(n):
    if n%4==0:
        if n%100== 0:
            if n%400==0:
                return 'Leap year'
            else:
                return 'Not a Leap year'
        else:
            return 'Not a Leap year'
    else:
        return 'Not a Leap year'
```

```python
n=int(input('enter a number : '))
check_year(n)
```

### Q5. Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?

```python
lst=[]
for i in range(1,10000):
    t=0
    for j in range(1,1000):
       
        if i%j==0:
            t=t+1
    if t==2:
        lst.append(i)    
```
